
# Problem 2 â€“ Coin Toss Simulation (Central Limit Theorem)

We simulate flipping a **fair coin** where:

- Heads = 1  
- Tails = 0  

We take random samples of size `n = 5, 10, 100`, compute their means, and repeat this **1000 times**.  
Finally, we display how the distribution of sample means approaches a **normal distribution** as sample size grows.


```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For reproducibility
np.random.seed(42)
```

## Simulation Function

```python

def simulate_coin_toss_means(n, trials=1000):
    """Simulate sample means for fair coin toss."""
    means = []
    for _ in range(trials):
        sample = np.random.choice([0,1], size=n, p=[0.5,0.5])
        means.append(sample.mean())
    return np.array(means)
```

## Run Simulations for n = 5, 10, 100

```python

sample_sizes = [5, 10, 100]
results = {n: simulate_coin_toss_means(n, trials=1000) for n in sample_sizes}
results
```

## Visualize the Distributions of Sample Means

```python

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, n in zip(axes, sample_sizes):
    sns.histplot(results[n], bins=20, kde=True, ax=ax, color="skyblue")
    ax.set_title(f"Distribution of Sample Means (n={n})")
    ax.axvline(0.5, color="red", linestyle="--", label="True Mean (0.5)")
    ax.legend()

plt.suptitle("Coin Toss Simulation â€“ Sample Mean Distributions", fontsize=16)
plt.show()
```


## Conclusion

- As the sample size **n increases** (5 â†’ 10 â†’ 100), the distribution of sample means becomes **tighter** and more closely resembles a **normal distribution** centered around 0.5.
- This illustrates the **Central Limit Theorem (CLT)**: regardless of the underlying distribution (Bernoulli in this case), the distribution of the sample mean approaches a normal distribution as `n` grows large.
